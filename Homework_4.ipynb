```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255.0
x_test /= 255.0
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
def create_model(num_filters=32, num_layers=2):
    model = Sequential()
    for i in range(num_layers):
        model.add(Conv2D(num_filters, (3, 3), padding='same', activation='relu', input_shape=x_train.shape[1:]))
        model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model
```

```python
def train_model(model, batch_size=32, epochs=20, learning_rate=0.01):
    adam = Adam(learning_rate=learning_rate)
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(x_test, y_test), shuffle=True)
    return model
```

```python
base_model = create_model()
base_model = train_model(base_model)
```

```python
wide_models = []
num_filters_list = [32, 64, 128]
for num_filters in num_filters_list:
    model = create_model(num_filters=num_filters)
    model = train_model(model)
    wide_models.append(model)
```

```python
deep_models = []
num_layers_list = [2, 4]
for num_layers in num_layers_list:
    model = create_model(num_layers=num_layers)
    model = train_model(model)
    deep_models.append(model)
```

```python
models = [base_model] + wide_models + deep_models
for i, model in enumerate(models):
    print(f"\nModel {i+1} summary:")
    model.summary()
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
    print(f"Test accuracy for model {i+1}: {test_acc}")
```
